```python
import controlsim as csm
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.size"] = 8
# plt.style.use("dark_background")
```

## Plant Definition
The plant is a SISO system represented as:
$$
\dot{x} = A x + B u
$$
$$
y = C x + D u
$$
Where:
- $A = -0.2$
- $B = 1$
- $C = 1$
- $D = 0$

```python
A = -0.2
B = 1
C = 1
D = 0
sys = ctrl.ss(A, B, C, D)
G = ctrl.ss2tf(sys)
dt = 0.01
plant = csm.ProcessDefinition(G, 0.0, dt=dt)
```

## Reference Model Definition
The reference model describes the desired closed-loop behavior:
$$
\dot{x}_m = A_m x_m + B_m r
$$
$$
y_m = C_m x_m + D_m r
$$
Where:
- $A_m = -2$
- $B_m = 2$
- $C_m = 1$
- $D_m = 0$

```python
Am = -2
Bm = 2
Cm = 1
Dm = 0
sysm = ctrl.ss(Am, Bm, Cm, Dm)
plant_ref = csm.ProcessDefinition(sysm, 0.0, dt=dt)
```

## Ideal Controller Parameters
For perfect tracking, the ideal controller gains are:
$$
K_x = \frac{A_m - A}{B}
$$
$$
K_r = \frac{B_m}{B}
$$

```python
K_x = (Am - A) / B
K_r = (Bm) / B
print(f"Ideal K_x: {K_x:.2f}\nIdeal K_r: {K_r:.2f}")
A_cl = A + B * K_x
print(f"Closed-loop A_cl: {A_cl:.2f}")
B_cl = B * K_r
print(f"Closed-loop B_cl: {B_cl:.2f}")

sys_cl = ctrl.ss(A_cl, B_cl, C, D)
G_cl = ctrl.ss2tf(sys_cl)
```

## Output Comparison
The following plots compare the plant output, reference model output, and ideal closed-loop output. Tracking error, control input, and adaptive gain evolution are also shown.

```python
t = np.arange(0, 10, dt)
R = np.ones_like(t)
plant.reset()
plant_ref.reset()
plant_cl = csm.ProcessDefinition(G_cl, 0.0, dt=dt)
plant_cl.reset()

for r in R:
    plant.step(r)
    plant_ref.step(r)
    plant_cl.step(r)

plt.figure()
plt.plot(plant.t, plant.y, label="Plant actual Output")
plt.plot(plant_ref.t, plant_ref.y, label="Desired Reference Model Output")
plt.plot(
    plant_cl.t, plant_cl.y, label="Ideal Closed-loop Output", linestyle=":", linewidth=3
)
plt.xlabel("Time")
plt.ylabel("Output")
plt.legend()
plt.show()
```

## MRAC Design Loop and Adaptive Law
The controller parameters $\hat{K}_x$ and $\hat{K}_r$ are updated using the adaptive law:
$$
\hat{K}_x= \Gamma\int_0^T  e B x ~ dt
$$
$$
\hat{K}_r= \Gamma\int_0^T e B r ~dt
$$
Where:
- $e = y - y_m$ (tracking error)
- $\Gamma$ is the adaptation gain

```python
t = np.arange(0, 100, dt)
R = np.ones_like(t)
x = 0
x_m = 0
K_xhat = 0
K_rhat = 0
e_h = []
G_f = ctrl.tf([1], [0.1, 1])
spf = csm.ProcessDefinition(G_f, 0.0, dt=dt)  # setpoint filter for reference input
spf.reset()
plant.reset()
plant_ref.reset()
Gamma = 2

K_x_h = np.array([])
K_r_h = np.array([])
u_h = np.array([])

R = np.sign(np.sin(2 * np.pi * 0.05 * t))  # Square wave reference input

for r in R:
    _, ref, t = spf.step(r)
    u = K_xhat * x + K_rhat * ref
    x, y, t = plant.step(u)
    x_m, y_m, t_m = plant_ref.step(ref)
    e = y - y_m

    # Adaptive law updates
    K_xhat += -Gamma * e * B * x * dt
    K_rhat += -Gamma * e * B * ref * dt

    # save the results
    e_h.append(e)
    K_x_h = np.append(K_x_h, K_xhat)
    K_r_h = np.append(K_r_h, K_rhat)
    u_h = np.append(u_h, u)
```

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 10))

axes[0].plot(plant.t, plant.y, label="Plant Output", linestyle="-", color="blue")
axes[0].plot(spf.t, spf.y, label="Filtered Reference Input", linestyle=":", color="red")
axes[0].plot(
    plant_ref.t,
    plant_ref.y,
    label="Reference Model Output",
    linestyle="--",
    color="black",
)
axes[0].set_xlabel("Time")
axes[0].set_ylabel("Output")
axes[0].legend()

axes[1].plot(plant.t[0 : len(e_h)], e_h, label="Tracking Error")
axes[1].set_xlabel("Time")
axes[1].set_ylabel(r"$e(t)$")
axes[1].legend()

axes[2].plot(plant.t[0 : len(u_h)], u_h, label=r"$u(t)$")
axes[2].set_xlabel("Time")
axes[2].set_ylabel(r"$u(t)$")
axes[2].legend()

axes[3].plot(plant.t[0 : len(K_x_h)], K_x_h, label=r"$\hat{K}_x$")
axes[3].axhline(K_x, color="red", linestyle="--", linewidth=1, label=r"Ideal $K_x$")
axes[3].plot(plant.t[0 : len(K_r_h)], K_r_h, label=r"$\hat{K}_r$")
axes[3].axhline(K_r, color="blue", linestyle="--", linewidth=1, label=r"Ideal $K_r$")
axes[3].set_xlabel("Time")
axes[3].set_ylabel(r"$K(t)$")
axes[3].legend()

plt.tight_layout()

plt.savefig("Figure_DMRAC.png", dpi=300)
plt.show()
```
